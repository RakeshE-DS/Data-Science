## **Background**


Voter turnout refers to the percentage of eligible voters who actually cast a ballot in a given election. In the United States, voter turnout has varied widely over the years, depending on a variety of factors such as the type of election, the level of enthusiasm among the electorate, and the accessibility of polls.
Historically, voter turnout in the United States has been relatively low compared to other democracies around the world. In the 2016 Presidential election, for example, voter turnout was around 58%, which is relatively low compared to other countries with similar democratic traditions.
Various factors can impact voter turnout. One major factor is the timing and format of elections. Additionally, the presence or absence of certain types of candidates or ballot initiatives can influence voter turnout. Candidates that generate high levels of enthusiasm or controversy often motivate voters to turn out in greater numbers.
Another factor that impacts voter turnout is the convenience and accessibility of polling locations. Long lines, difficulty in obtaining absentee ballots, and issues with voter registration can all discourage people from participating in elections.
In recent years, various proposals have been put forth aimed at increasing voter turnout, such as making election day a holiday, expanding early voting, and simplifying the registration process. These proposals seek to ease the logistical and practical barriers that can discourage people from participating in democratic processes.

## **Historical trends**

Voter turnout in the US has varied widely over the years, with some elections seeing high levels of participation and others seeing low levels. One of the earliest recorded presidential elections with relatively high turnout was in 1824, which had a turnout rate of 27.9%. This was partly due to the expansion of voting rights and increased participation among white male citizens. However, voter turnout declined in subsequent presidential elections, reaching a low point in 1924 with a turnout rate of just 48.9%.
Voter turnout rates began to increase again in the 1960s, following the passage of the Voting Rights Act in 1965, which eliminated many of the legal barriers to voting that had existed in some states. The 1964 presidential election saw a turnout rate of 61.9%, which was the highest since 1908. Since then, voter turnout rates have remained relatively stable, with some fluctuations depending on the election year and other factors. Presidential elections generally have higher turnout rates than midterm elections, which are held in years when there are no presidential candidates on the ballot.
However, despite some fluctuations, the trend over the past several decades has been towards lower voter turnout. In the 1996 presidential election, for example, just 49% of eligible voters cast a ballot, and in the 2014 midterm election, only 36.4% of eligible voters did so. This trend towards lower voter turnout is troubling for a variety of reasons, including the fact that it may indicate a decline in overall civic engagement and an erosion of democratic legitimacy. There are many proposals that have been put forth to address this trend, such as expanding early voting and making election day a holiday to make it easier for people to participate.

#### **Function Declarations for Census Data Collection**

The provided code includes three functions: rename_columns_with_variable_codes, clean_variable_definition, and get_variable_definitions.

 The rename_columns_with_variable_codes function renames the columns of a pandas dataframe using a dictionary of variable codes and new names, the clean_variable_definition function cleans up the raw census definition strings, and the get_variable_definitions function retrieves the cleaned definitions of census variables from a JSON file and returns them as a pandas dataframe.

 These functions would be useful for working with census data and improving the readability and understandability of the variable names and definitions in the dataset.

The Census web services hold 'variables.json' APIs that contain the data dictionaries. As such we created functions to extract the json from the web services, rename the variables (from the census label that isn't descriptive) and finally clean any definition strings that get returned. For example, often the definition strings include many '!!' that if removed can improve the human readability of the strings. At present these functions exist in our project, but are not used to generate the markdown table of dictionaries below. This allows for future expansion of the codebase.

```python
def rename_columns_with_variable_codes(df, variable_codes):
    """
    df: a data frame to rename columns
    variable_codes: dictionary maping the new names to the original column names
    returns: modified DataFrame with new column names
    """

    # -- The api variable key names and new dataframe column names dictionary
    reversed_variable_codes = {code: name for name, code in variable_codes.items()}
    df = df.rename(columns=reversed_variable_codes)

    return df
```

```python
def clean_variable_definition(variable_definition):
    """ variable_definition: raw census definition
      returns: cleaned string for census definition"""

    levels = variable_definition.split("!!")
    if levels[0] == "Estimate":
        levels[0] = "Est"
    return " - ".join(levels)
```

```python
def get_variable_definitions(vars_url):
    """
     vars_url: url for census endingin variables.json
     returns: definition fields for variables
    """

    response = requests.get(vars_url)
    if response.status_code != 200:
        print(f"Error: Failed to get variable definitions for {vars_url}")
        return None

    variables_dict = response.json()['variables']
    variables_list = []
    for code, definition in variables_dict.items():
        variables_list.append((code, clean_variable_definition(definition['label'])))

    variables_df = pd.DataFrame(variables_list, columns=['code', 'definition'])
    return variables_df
```

```python

def get_census_data(api_url=None, req_key_names=None, out_df_name=None):
    """
      get cencus data from acs5 API url and store the data in the dataset

      url: API endpoint
      params: parameters for the API request
      returns: df with the census data
    """

    # Send a GET request to the Census API with the specified URL and parameters
    api_response = requests.get(api_url, params=req_key_names)

    try:
        # Try to parse the JSON response
        data = api_response.json()

        # Remove the first element of the JSON response, which contains the headers
        headers = data.pop(0)

        # Create a DataFrame using the data and headers
        df = pd.DataFrame(data, columns=headers)
    except Exception as e:
        # If an exception occurs (e.g., invalid JSON response), print the error and response
        print(f"Error: {e}")
        print(f"Response: {response.text}")

        # Return None to indicate that the request failed
        return None

    # Return the DataFrame with the census data
    return df
```

```python
def get_census_data_for_presidential_years(years, census_api_key, api_variable_keys_dict):
    """
      to get the county level census data for the given presidential election years
    """

    all_years_data = []

    # loop for each year and get data from the API
    for year in years:
        census_url = f"https://api.census.gov/data/{year}/acs/acs5"

        # make a comma seperated list of the codes from json
        get_vars = ','.join(
            [f"{code}" for code in api_variable_keys_dict.values()])

        # put the list of variables in the params to send to the api
        census_params = {
            'get': f"NAME,{get_vars}",
            'for': 'county:*',
            'key': census_api_key
        }

        # get the dataframe created with the data from the census API
        census_df = get_census_data(census_url, census_params)

        if census_df is not None:
            # Rename columns with the column names given in the api_variable_keys_dict
            census_df = rename_columns_with_variable_codes(census_df, api_variable_keys_dict)

            # Add year column
            census_df['year'] = year

            all_years_data.append(census_df)

    # add all of the years data to a single dataframe
    all_years_data_df = pd.concat(all_years_data, ignore_index=True)

    # Contstruct the GEOID on the census data
    all_years_data_df['GEOID'] = all_years_data_df['state'] + all_years_data_df['county']
    all_years_data_df.set_index('GEOID', inplace=True)

    return all_years_data_df
```

### Data Visualization  
_Plots and Coropleths_

In an effort to reduce code redundancy, the plot_county_choropleth() function is takes in some parameters and generates a choropleth map of the United States displaying the specified feature by county for the specified year.

The function loads a shapefile containing the county boundary information of the United States, and merges the specified feature data with the shapefile data. It then filters out non-continental US counties, converts the specified feature column to integer, sets the minimum and maximum values for the colorbar, and creates a choropleth map using the merged data. Finally, the function adjusts the colorbar size, sets the title and axis labels, and returns the Matplotlib axis object. This function is useful for exploring the distribution of a feature across counties in the United States.

```python
def plot_county_choropleth(df, feature, year, ax):
    '''Pass this function a data frame WITH A 5digit GEOID and a feature, and it will create a coropleth '''

    # Load county shape file and data
    shp_file_path = files_path_prefix + 'MLPP23 Group 1 Project/data/cb_2019_us_county_20m/cb_2019_us_county_20m.shp'
    county_voters_shape_file = gpd.read_file(shp_file_path)

    # Manipulate the data a bit.. .merge data set to shp and remove non-contenental us
    county_voters_shape_file = county_voters_shape_file.merge(df, on='GEOID', how='left')
    county_voters_shape_file = county_voters_shape_file[(county_voters_shape_file['year'] == year)] # exclude Hawaii
    county_voters_shape_file = county_voters_shape_file[(county_voters_shape_file['STATEFP'] != '15')] # exclude Hawaii
    county_voters_shape_file = county_voters_shape_file[(county_voters_shape_file['STATEFP'] != '02')] # exclude Alaska
    county_voters_shape_file = county_voters_shape_file[(county_voters_shape_file['STATEFP'] != '72')] # exclude Puerto Rico
    county_voters_shape_file = county_voters_shape_file[(county_voters_shape_file['STATEFP'] != '66')] # exclude Guam
    county_voters_shape_file = county_voters_shape_file[(county_voters_shape_file['STATEFP'] != '60')] # exclude American Samoa
    county_voters_shape_file = county_voters_shape_file[(county_voters_shape_file['STATEFP'] != '78')] # exclude US Virgin Islands
    county_voters_shape_file = county_voters_shape_file[(county_voters_shape_file['STATEFP'] != '69')] # exclude Northern Mariana Islands

    # Filter for the specified year and exclude Hawaii
    county_voters_shape_file_filtered = county_voters_shape_file # don't include hawaii because the map gets too small

    # Convert the specified feature column to integer
    county_voters_shape_file_filtered[feature] = county_voters_shape_file_filtered[feature].fillna(0).astype(float)

    # Set vmin and vmax values for colorbar
    my_vmin = county_voters_shape_file_filtered[feature].min()
    my_vmax = county_voters_shape_file_filtered[feature].max()

  # Create figure and axes
    if ax is None:
        fig, ax = plt.subplots(figsize=(20, 20))
    else:
        fig = ax.get_figure()

    # Plot choropleth map
    county_voters_shape_file_filtered.plot(column=feature, vmin=my_vmin, vmax=my_vmax, legend=True, ax=ax, linewidth=0.05)

    # Adjust colorbar size
    cb = ax.get_figure().get_axes()[1]
    cb.set_position([0.95, 0.2, 0.02, 0.6])
    cb.set_label(feature)

    # Set title and axis labels
    ax.set_title(f'{feature} by County {year}', fontsize=20)
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')

    return ax
```

Below we combine age groups in a dataframe and calculate the percentages of each column, to get the percentage of eligible voters who belong to a particular age group. The purpose of this transformation is to reduce the impact of correlation among age groups on a machine learning model that you see in the above correlation matrix.

The code combines age groups 18-29 into one group, age groups 30-59 into another group, and age groups 60+ into the third group. After combining the age groups, the code calculates the total voting population by adding up the combined age groups. Then, it calculates the percentage of eligible voters who belong to each age group by dividing the combined age group count by the total voting population and multiplying by 100. The resulting percentages are stored in new columns in the dataframe, which will be used for features in the model moving foward. By reducing the correlation among age groups, the hope is that the machine learning model will be less biased and more accurate in its predictions.

The code does not explicitly perform normalization of the data between counties with higher populations vs. lower population counties (ie: scaling the values to a specific range, such as between 0 and 1, to make them directly comparable across different scales), the calculation of percentages for each age group relative to the total voting population can indirectly help to normalize the data across counties with different population sizes. When calculating the percentages of eligible voters in the age groups relative to the total voting population, the size of the county population is taken into account. Therefore, when comparing percentages across counties, the population size of each county is considered. By calculating the percentages of each age group relative to the total voting population, the code indirectly helps to normalize the data across counties with different population sizes, which proved to be beneficial for analysis and modeling purposes.

```python
def combine_cols_get_percent(input_df, req_voter_turnout_cols):
    # combining age columns to reduce correlation

    age_18_to_29 = input_df['age_18_to_19'] +  input_df['age_25_to_29'] + \
                  input_df['age_20_years'] + input_df['age_21_years'] + \
                  input_df['age_22_to_24'] \

    age_30_to_59 = input_df['age_30_to_34'] + input_df['age_35_to_39'] + input_df['age_40_to_44'] + \
                  input_df['age_45_to_49'] + input_df['age_50_to_54'] + input_df['age_55_to_59']
    age_above_60 = input_df['age_60_to_61'] + input_df['age_62_to_64'] + input_df['age_65_to_66'] + \
                  input_df['age_67_to_69'] + input_df['age_70_to_74'] + input_df['age_75_to_79'] + \
                  input_df['age_80_to_84']

    # calculate the percentages of each column, to get the percentage of eigible voters, who belongs to a particular age group

    # **we are calculating percentages only relative to eligible voters age groups(age greater than 18)
    total_voting_population = age_18_to_29 + age_30_to_59 + age_above_60
    input_df['age_18_to_29_percent'] = (age_18_to_29 / total_voting_population) * 100
    input_df['age_30_to_59_percent'] = (age_30_to_59 / total_voting_population) * 100
    input_df['age_above_60_percent'] = (age_above_60 / total_voting_population) * 100

    # combining education columns to reduce correlation
    studied_degree_or_more = input_df['completed_masters_degree'] + input_df['completed_associate_degree'] + \
                            input_df['completed_professional_school_degree'] + input_df[
                                'completed_doctoral_degree'] + input_df['completed_bacheolers_degree']

    studied_less_than_degree = input_df['completed_high_school_diploma'] + input_df[
        'no_schooling_completed'] + input_df['attended_college_no_degree']

    total_educated_people = studied_degree_or_more + studied_less_than_degree

    # calculate the percentages of each column, to get the percentage of eigible voters, who belongs to a particular education level
    input_df['studied_degree_or_more_percent'] = (studied_degree_or_more / total_educated_people) * 100
    input_df['studied_less_than_degree_percent'] = (studied_less_than_degree / total_educated_people) * 100

    if 'below_18_years_age' in list(input_df.columns):
      # drop the original age columns, as we will use only age percentage groups, from now
      input_df.drop(columns=age_column_names, inplace=True)

    # drop the education columns, as we will use only age percentage groups, from now
    input_df.drop(columns=education_cols_list, inplace=True)

    # dropping age less than 18 years columns, as these people are not voters and are not contributing for our end goal, we have taken this column, for visualization purpose only
    input_df.drop('below_18_years_age', axis=1, inplace=True)

    cols_not_needed = education_cols_list + age_column_names + ['above_poverty_level_female', 'above_poverty_level_male']

    # remove all the education variables from our required variables list also
    req_voter_turnout_cols = [variable_name for variable_name in req_voter_turnout_cols if
                                  variable_name not in cols_not_needed]

    if 'below_18_years_age' in req_voter_turnout_cols:
      req_voter_turnout_cols.remove('below_18_years_age')

    # voter_turnout_df['above_poverty_level'] = voter_turnout_df['above_poverty_level_male'] + voter_turnout_df['above_poverty_level_female']
    # voter_turnout_df.drop(columns=['above_poverty_level_female', 'above_poverty_level_male'], inplace=True)


    # adding the new age and education percentage columns to req_voter_turnout_cols list
    req_voter_turnout_cols.extend(
        ['voter_turnout_percent'] + ['age_18_to_29_percent', 'age_30_to_59_percent', 'age_above_60_percent',
                                    'studied_less_than_degree_percent', 'studied_degree_or_more_percent',
                                    #  'above_poverty_level'
                                     ])


    return req_voter_turnout_cols

req_voter_turnout_cols = combine_cols_get_percent(voter_turnout_df, req_voter_turnout_cols)
```

1. Mean Absolute Error (MAE):
The MAE is a measure of the average absolute difference between the predicted values and the actual values. It represents the average absolute error of the model's predictions. It is calculated as the mean of the absolute difference between the predicted and actual values:

    
    MAE = (1/n) * sum(abs(y_actual - y_predicted))


where y_actual are the actual values and y_predicted are the predicted values. The MAE value ranges between 0 and infinity, with lower values indicating better performance.
2. Root Mean Square Error (RMSE):
The RMSE is a measure of the average magnitude of the error between the predicted values and the actual values. It represents the square root of the average squared difference between the predicted and actual values. It is calculated as:


    RMSE = sqrt((1/n) * sum((y_actual - y_predicted)^2))


where y_actual are the actual values and y_predicted are the predicted values. The RMSE value ranges between 0 and infinity, with lower values indicating better performance.
3. R squared (R2):
R2 is a statistical measure that represents the proportion of the variance in the dependent variable (y) that is explained by the independent variables (x) in the model. It is calculated as:


    R2 = 1 - (sum(y_actual - y_predicted)^2) / (sum(y_actual - mean(y_actual))^2)


where y_actual are the actual values, y_predicted are the predicted values, and mean(y_actual) is the mean of the actual values. The R2 value ranges between 0 and 1, with higher values indicating better performance. An R2 value of 1 indicates that the model explains all the variability of the dependent variable, while an R2 value of 0 indicates that the model does not explain any variability of the dependent variable.

#why different models have different feature importances

The feature importances for tree-based models like decision trees and random forests can be different because they use different methods to determine feature importance.

In a decision tree, feature importance is determined by the reduction in impurity (such as Gini impurity or entropy) caused by splitting on that feature. The more a feature reduces impurity, the higher its importance.

In a random forest, feature importance is determined by averaging the impurity reduction of that feature over all the trees in the forest. This provides a more robust estimate of feature importance as it reduces the impact of overfitting to any individual tree.

Additionally, random forests can capture interactions between features, while decision trees only consider one feature at a time. This means that the feature importances for random forests may be influenced by interactions between features, whereas decision trees can only capture the importance of each feature in isolation.




In the case of your SGDRegressor model, it is possible that the algorithm is placing more importance on the "Voter_participation_count" feature because it has a strong correlation with the target variable, which is the voter turnout percentage. However, it is also possible that this feature is not actually the most important one for predicting voter turnout percentage and that the model is overfitting to the training data.
